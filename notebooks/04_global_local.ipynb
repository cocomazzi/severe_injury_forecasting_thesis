```python
from src.features import compute_naics_mix

naics_mix = compute_naics_mix(
    df=df,
    naics_col="Primary NAICS",
    date_col="EventDate",
    state_col="State",
    freq="MS",
    top_k=10,
    out_date_col="Date"
)
```
